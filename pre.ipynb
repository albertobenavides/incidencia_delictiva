```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1997_2017 = pd.read_csv('datos/IDEFC_jun2022.csv', encoding = "ISO-8859-1")
df_1997_2017
```

```python
df_1997_2017.MODALIDAD.unique()
```

```python
df_1997_homicidios = df_1997_2017[df_1997_2017['MODALIDAD'] == 'HOMICIDIOS']
df_1997_homicidios
```

```python
df_1997_homicidios.TIPO.unique()
```

```python
df_1997_dolosos = df_1997_homicidios[df_1997_homicidios.TIPO == 'DOLOSOS']
df_1997_dolosos = df_1997_dolosos.drop(columns = 'INEGI')
df_1997_dolosos
```

```python
gp_1997_dolosos = df_1997_dolosos.groupby(['AÑO', 'ENTIDAD'])
df_1997_dolosos_sum = gp_1997_dolosos.sum()
df_1997_dolosos_sum = df_1997_dolosos_sum.reset_index()
df_1997_dolosos_sum
```

```python
c = ['year', 'state']
c.extend(list(range(1, 13)))
c = [str(i) for i in c]
c
```

```python
df_1997_dolosos_sum.columns = c
df_1997_dolosos_sum
```

```python
t = df_1997_dolosos_sum.melt(
  id_vars=["year", "state"], 
        var_name="month", 
        value_name="value"
)
t
```

```python
df_dolosos = pd.pivot_table(t, values='value', index=['year', 'month'], columns=['state']).reset_index()
df_dolosos
```

```python
# https://stackoverflow.com/a/37354256
from pandas.tseries.offsets import MonthEnd
pd.to_datetime(df_dolosos['year'].astype(str) + df_dolosos['month'], format="%Y%m") + MonthEnd(0)
```

```python
df_dolosos.index = pd.to_datetime(df_dolosos['year'].astype(str) + df_dolosos['month'], format="%Y%m") + MonthEnd(0)
df_dolosos = df_dolosos.drop(columns = ['year', 'month'])
df_dolosos
```

```python
df_dolosos.to_csv('datos/1997.csv')
```

```python
df_final = df_dolosos.sum(axis = 1).sort_index()
```

```python
plt.figure(figsize = (20, 4))
plt.plot(df_final)
plt.show()
```

```python
df_2015_2022 = pd.read_csv('datos/IDEFC_NM_ago22.csv', encoding = "ISO-8859-1")
df_2015_2022
```

```python
df_2015_2022['Tipo de delito'].unique()
```

```python
df_2015_homicidios = df_2015_2022[df_2015_2022['Tipo de delito'] == 'Homicidio']
df_2015_homicidios
```

```python
df_2015_homicidios['Subtipo de delito'].unique()
```

```python
df_2015_dolosos = df_2015_homicidios[df_2015_homicidios['Subtipo de delito'] == 'Homicidio doloso']
df_2015_dolosos
```

```python
gp_2015_dolosos = df_2015_dolosos.groupby(['Año', 'Entidad'])
df_2015_dolosos_sum = gp_2015_dolosos.sum()
df_2015_dolosos_sum = df_2015_dolosos_sum.reset_index()
df_2015_dolosos_sum = df_2015_dolosos_sum.drop(columns = 'Clave_Ent')
df_2015_dolosos_sum
```

```python
c = ['year', 'state']
c.extend(list(range(1, 13)))
c = [str(i) for i in c]
c
```

```python
df_2015_dolosos_sum.columns = c
df_2015_dolosos_sum
```

```python
t = df_2015_dolosos_sum.melt(
  id_vars=["year", "state"], 
        var_name="month", 
        value_name="value"
)
t
```

```python
df_2015_dolosos = pd.pivot_table(t, values='value', index=['year', 'month'], columns=['state']).reset_index()
df_2015_dolosos
```

```python
from pandas.tseries.offsets import MonthEnd
pd.to_datetime(df_2015_dolosos['year'].astype(str) + df_2015_dolosos['month'], format="%Y%m") + MonthEnd(0)
```

```python
df_2015_dolosos.index = pd.to_datetime(df_2015_dolosos['year'].astype(str) + df_2015_dolosos['month'], format="%Y%m") + MonthEnd(0)
df_2015_dolosos = df_2015_dolosos.drop(columns = ['year', 'month'])
df_2015_dolosos
```

```python
df_2015_dolosos.to_csv('datos/2015.csv')
```
